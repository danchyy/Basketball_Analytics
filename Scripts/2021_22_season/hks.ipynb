```python
import json
import pandas as pd
import os
```

```python
from bs4 import BeautifulSoup
```

```python
url = "https://hosted.dcd.shared.geniussports.com/embednf/HKS/en/schedule?&iurl=https%3A%2F%2Fwww.hks-cbf.hr%2Fstatistika%2F&_cc=1&_lc=1&_nv=1&_mf=1"
```

```python
resp = requests.get(url)
```

```python
json_data = resp.json()
```

```python
soup = BeautifulSoup(json_data['html'])
```

```python
items = soup.find_all("a", attrs={'class': 'detail-r-link'})
```

```python
resp = requests.get("https://fibalivestats.dcd.shared.geniussports.com/data/1927416/data.json")
```

```python
total_items = len(items)
for index, item in enumerate(items):
    print(f"{index}/{total_items}")
    href_link = item.attrs['href']
    if 'summary' not in href_link:
        continue
    splitted = href_link.split("%2F")
    url = f"https://fibalivestats.dcd.shared.geniussports.com/data/{splitted[-2]}/data.json"
    #url = f"https://fibalivestats.dcd.shared.geniussports.com/u/HKS/"
    print(url)
    try:
        json_data = requests.get(url).json()
    except Exception:
        print("Not available")
        continue
    with open(f'hr_premier_pbp/{splitted[-2]}.json', 'w') as f:
        json.dump(json_data, f)
    time.sleep(0.5)
    
```

```python
pbp_dfs = []
for item in os.listdir('hr_premier_pbp/'):
    full_path = f'hr_premier_pbp/{item}'
    with open(full_path) as f:
        json_data = json.load(f)
    cur_pbp = json_data['pbp']
    df = pd.DataFrame(cur_pbp)
    df = df.sort_values(by='actionNumber')
    df.loc[:, 'game_id'] = item.split(".")[0]
    first_team = json_data['tm']['1']['name']
    second_team = json_data['tm']['2']['name']
    team_numbers = df.tno.tolist()
    teams = []
    for team_no in team_numbers:
        if team_no == 1:
            teams.append(first_team)
        elif team_no == 2:
            teams.append(second_team)
        else:
            teams.append("")
    df.loc[:, 'team'] = teams
    pbp_dfs.append(df)
```

```python
full_pbp_dfs = pd.concat(pbp_dfs)
```
