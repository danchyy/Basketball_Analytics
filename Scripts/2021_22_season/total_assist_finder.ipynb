```python
import pandas as pd
import json
from nba_api.stats.static.players import find_players_by_full_name
from nba_api.stats.static.teams import find_teams_by_full_name
```

```python
cavs_id = find_teams_by_full_name("Cleveland Cavaliers")[0]['id']
```

```python
cavs_id
```

```python
rubio_id = find_players_by_full_name("Ricky Rubio")
```

```python
rubio_id = rubio_id[0]['id']
```

```python
seasons = [f"{item}-{str(item+1)[-2:]}" for item in range(2011, 2022)]
```

```python
seasons
```

```python
dfs = []
for season in seasons:
    df = pd.read_csv(f'pbp_data/{season}.csv')
    dfs.append(df)
```

```python
full_df = pd.concat(dfs)
```

```python
rubio_assists = full_df.loc[(full_df.EVENTMSGTYPE == 1) & (full_df.PLAYER2_ID == rubio_id)]
```

```python
rubio_ast = rubio_assists.groupby(['PLAYER1_NAME', 'PLAYER1_ID']).size().to_frame().reset_index().rename(columns={'PLAYER1_NAME': 'PLAYER NAME', 0: 'TOTAL ASSISTS'})
```

```python
rubio_ast.sort_values(by='TOTAL ASSISTS', ascending=False)
```

```python
rubio_ast.sort_values(by='TOTAL ASSISTS', ascending=False).reset_index().loc[:, ['PLAYER NAME', 'TOTAL ASSISTS']]
```
