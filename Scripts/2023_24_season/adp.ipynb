```python
import requests
import json
import pandas as pd
```

```python
url = "https://www.rotowire.com/basketball/tables/adp.php?pos=ALL&season=2023"
```

```python
resp = requests.get(url)
```

```python
current_adp = pd.DataFrame(resp.json())
```

```python
past_adp = pd.DataFrame(requests.get("https://www.rotowire.com/basketball/tables/adp.php?pos=ALL&season=2022").json())
```

```python
subset_current = current_adp.loc[:, ['player', 'yahoo']]
```

```python
subset_past = past_adp.loc[:, ['player', 'yahoo']]
```

```python

merged_df = pd.merge(left=subset_past, right=subset_current, on='player', suffixes=['_22_23', '_23_24'])
```

```python
merged_df.yahoo_22_23 = merged_df.yahoo_22_23.replace('-', '160.00')
```

```python
merged_df.yahoo_23_24 = merged_df.yahoo_23_24.replace('-', '160.00')
```

```python
merged_df.yahoo_22_23 = merged_df.yahoo_22_23.astype(float)
merged_df.yahoo_23_24 = merged_df.yahoo_23_24.astype(float)
```

```python
merged_df.loc[:, 'change'] = merged_df.yahoo_22_23 - merged_df.yahoo_23_24
```

```python
top_30 = merged_df.loc[(merged_df.yahoo_22_23 < 150)].sort_values(by='change', ascending=False).iloc[:50].reset_index()
```

```python
print(top_30.loc[:, top_30.columns[1:]].to_markdown())
```
