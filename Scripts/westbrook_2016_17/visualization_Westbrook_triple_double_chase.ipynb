```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
westbrook_stats = pd.read_csv('../../Data/westbrook_2016_17/westbrook_2016_17.csv')
```

```python
westbrook_stats
```

```python
def truncate_float(number):
    return "%.2f" % number
```

```python
pts, rbs, ast = westbrook_stats.PTS, westbrook_stats.TRB, westbrook_stats.AST
pts_average, rbs_average, ast_average = [], [], []
for i in range(len(pts)):
    if i == 0:
        pts_average.append(truncate_float(pts[i]))
        rbs_average.append(truncate_float(rbs[i]))
        ast_average.append(truncate_float(ast[i]))
    elif i == 80:
        pts_average.append(truncate_float(pts[:(i-1)].mean()))
        rbs_average.append(truncate_float(rbs[:(i-1)].mean()))
        ast_average.append(truncate_float(ast[:(i-1)].mean()))
    else:
        pts_average.append(truncate_float(pts[:i].mean()))
        rbs_average.append(truncate_float(rbs[:i].mean()))
        ast_average.append(truncate_float(ast[:i].mean()))

westbrook_stats.loc[:, "AvgPTS"] = pts_average
westbrook_stats.loc[:, "AvgTRB"] = rbs_average
westbrook_stats.loc[:, "AvgAST"] = ast_average
```

```python
westbrook_stats
```

```python
plt.style.available
```

```python
plt.style.use('seaborn')
```

```python
plt.figure(figsize=(12, 12))

plt.plot(westbrook_stats.Rk.astype(int), westbrook_stats.AvgPTS.astype(float), label="Points per game", color="#c5423d")
plt.plot(westbrook_stats.Rk.astype(int), westbrook_stats.AvgTRB.astype(float), label="Rebounds per game", color="#4341c2")
plt.plot(westbrook_stats.Rk.astype(int), westbrook_stats.AvgAST.astype(float), label="Assist per game", color="#40c23b")
plt.xlabel("Game in season")
plt.ylabel("PTS/TRB/AST per game")

plt.title("Russell Westbrook's averages during his 2016-17 NBA season")

plt.legend()
plt.xlim(0, 84)
```

```python
westbrook_stats = westbrook_stats.drop(westbrook_stats.index[80]) # dropping duplicate game
```

```python
pts_season, trb_season, ast_season = westbrook_stats.PTS.mean(), westbrook_stats.TRB.mean(), westbrook_stats.AST.mean()
```

```python
pts, rbs, ast = westbrook_stats.AvgPTS, westbrook_stats.AvgTRB, westbrook_stats.AvgAST
pts_change, rbs_change, ast_change = [], [], []
for i in pts.index.tolist():
    pts_change.append(truncate_float(float(pts[i]) - pts_season))
    rbs_change.append(truncate_float(float(rbs[i]) - trb_season))
    ast_change.append(truncate_float(float(ast[i]) - ast_season))

westbrook_stats.loc[:, "PTSChange"] = pts_change
westbrook_stats.loc[:, "TRBChange"] = rbs_change
westbrook_stats.loc[:, "ASTChange"] = ast_change
```

```python
westbrook_stats
```

```python
plt.figure(figsize=(16, 14))

plt.plot(westbrook_stats.Rk.astype(int), westbrook_stats.PTSChange.astype(float), label="Difference between points' average through n-th game and season average of " + truncate_float(pts_season), color="#c5423d")
plt.plot(westbrook_stats.Rk.astype(int), westbrook_stats.TRBChange.astype(float), label="Difference between rebounds' average through n-th game and season average of " + truncate_float(trb_season), color="#4341c2")
plt.plot(westbrook_stats.Rk.astype(int), westbrook_stats.ASTChange.astype(float), label="Difference between assists' average through n-th game and season average of" + truncate_float(ast_season), color="#40c23b")

plt.xlabel("Game in season", size=14)
plt.ylabel("Change in average PTS/TRB/AST through that game in relation to whole season", size=14)

plt.title("Russell Westbrook's averages through x-th game in relation to whole season averages during his 2016-17 NBA season", size=14)

plt.legend()
```
